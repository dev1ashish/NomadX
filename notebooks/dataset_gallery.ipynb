# Dataset Gallery — side-by-side view

Every plot in `images/` laid out inline so you can scroll through the whole dataset in one place. No code, no execution — just markdown referencing the PNGs generated by `dataset_visualization.ipynb`.

**Layout:**
1. Six dataset-level summary plots (a 2×3 grid).
2. Then every file, grouped by class → subclass, with its four diagnostic plots side by side (overlay · heatmap · spatial · mean ± band).

If images don't render, run `dataset_visualization.ipynb` first to regenerate `images/`.
